```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt         # to generate plots
```

## 2. Load IOD SSTA Series

```python
#loading and assigning name to npzfile
npzfile = np.load('/content/drive/MyDrive/Project Learnings/Data/ssta.iod.21y.npz')
npzfile.files
```

```python
#loads SST time series data from 'ssta_series' array within the 'npzfile' file
ssta_series = npzfile['ssta_series']
ssta_series.shape
```

## 3. Have a quick plot

```python
plt.plot(ssta_series)
```

## 4. Make it  Beautiful

### 4.1 Add Year ticks and grid lines, etc.

More info can be found from https://matplotlib.org/users/pyplot_tutorial.html

```python
fig = plt.figure(figsize=(15, 6))
ax = fig.add_subplot(111)

plt.plot(ssta_series, 'g-', linewidth=2)

plt.xlabel('Years')
plt.ylabel('[$^oC$]')
plt.title('IOD SSTA 21-year (1979-1999)', fontsize=12)

ax.set_xlim(0,300)
ax.set_ylim(-2,2)
ax.set_xticklabels(range(1979,1999,1*4))
ax.axhline(0, color='r')

plt.grid(True)
ax.autoscale_view()

plt.savefig('/content/drive/MyDrive/Project Learnings/Data/ssta_series_21y.png')
```

### 4.2 More professional

Just like the image from https://www.esrl.noaa.gov/psd/enso/mei/

```python
fig = plt.figure(figsize=(15, 6))
ax = fig.add_subplot(111)

xtime       = np.linspace(1,252,252)
ssta_series = ssta_series.reshape((252))

ax.plot(xtime, ssta_series, 'black', alpha=1.00, linewidth=2)
ax.fill_between(xtime, 0., ssta_series, ssta_series> 0., color='red', alpha=.75)
ax.fill_between(xtime, 0., ssta_series, ssta_series< 0., color='blue',  alpha=.75)

plt.xlabel('Years')
plt.ylabel('[$^oC$]')
plt.title('IOD SSTA 21-year (1979-1999)', fontsize=12)

ax.set_xlim(xtime[0], xtime[-1])
ax.set_ylim(-2, 2)
ax.set_xticklabels(range(1979,2000,1*4))

plt.grid(True)
ax.autoscale_view()
```

**Key Observations:**

* **Fluctuations:** The plot exhibits significant fluctuations in SST anomalies, indicating a high degree of variability in the Indian Ocean Dipole (IOD) over the 21-year period.
* **Positive and Negative Anomalies:** The presence of both positive and negative anomalies suggests that the IOD has experienced both warm and cool phases during this time.
* **Periodic Patterns:** There appear to be some periodic patterns, potentially indicating cyclical variations related to climate modes like ENSO or other factors.
* **Extreme Events:** The plot shows a few instances of particularly strong positive or negative anomalies, suggesting the potential for significant impacts on regional climate and weather patterns.

**Conclusions:**

Based on these observations, we can draw the following conclusions:

* **IOD Variability:** The IOD is a highly variable climate phenomenon, with significant fluctuations in SST anomalies occurring over the 21-year period.
* **Impact on Regional Climate:** The IOD's influence on regional climate is likely substantial, given the observed extreme events and potential linkages to other climate modes.
* **Importance of Monitoring:** Continued monitoring of IOD variability is crucial for understanding its impacts on weather patterns, ecosystems, and human activities in the Indian Ocean region.
* **Climate Modeling:** Incorporating the IOD into climate models can improve predictions of regional climate variability and potential impacts.
